```python
import numpy as np
import pandas as pd
import keras
print(keras.__version__)
from tensorflow.keras import datasets     #MNIST
(x_train,y_train),(x_test, y_test)=datasets.mnist.load_data()

# Define image dimensions and reshape the data
width, height = 28, 28
x_train = x_train.reshape(x_train.shape[0], height, width, 1)
x_test = x_test.reshape(x_test.shape[0], height, width, 1)

# Train-validation split
from sklearn.model_selection import train_test_split
x_train, x_val, y_train, y_val = train_test_split(x_train, y_train, test_size=0.1)    # val = validation

# Normalize the data
x_train = (x_train - x_train.mean()) / x_train.std()
x_val = (x_val - x_val.mean()) / x_val.std()
x_test = (x_test - x_test.mean()) / x_test.std()

# Convert labels to one-hot encoding
num_labels = 10  # Since MNIST has 10 digits
y_train = keras.utils.to_categorical(y_train, num_classes=num_labels)
y_val = keras.utils.to_categorical(y_val, num_classes=num_labels)
y_test = keras.utils.to_categorical(y_test, num_classes=num_labels)

# Ensure the shapes are correct
print(f'y_train shape: {y_train.shape}')  # Should be (54000, 10)
print(f'y_val shape: {y_val.shape}')      # Should be (6000, 10)
print(f'y_test shape: {y_test.shape}')    # Should be (10000, 10)

from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout, Flatten, Conv2D, MaxPooling2D, AveragePooling2D

## LENET5 Model
model = Sequential()

# Feature Extraction (LeNet-5 architecture)
model.add(Conv2D(6, kernel_size=(5,5), strides=(1,1), activation='tanh', input_shape=(28, 28, 1)))
model.add(AveragePooling2D(pool_size=(2,2), strides=(1,1)))
model.add(Conv2D(16, kernel_size=(5,5), strides=(1,1), activation='tanh'))
model.add(AveragePooling2D(pool_size=(2,2), strides=(2,2)))
model.add(Conv2D(120, kernel_size=(5,5), strides=(1,1), activation='tanh'))

# Fully connected layers
model.add(Flatten())
model.add(Dense(84, activation='tanh'))
model.add(Dense(num_labels, activation='softmax'))

model.summary()

# Compile the model
model.compile(loss=keras.losses.categorical_crossentropy, optimizer='adam', metrics=['accuracy'])

# Train the model
history = model.fit(x_train, y_train, batch_size=32, epochs=20, verbose=1, validation_data=(x_val, y_val))
```

```python
# Train the model (this will define the input and output tensors)
history = model.fit(x_train, y_train, batch_size=32, epochs=1, verbose=1, validation_data=(x_val, y_val))  # Use 1 epoch for quicker training
```

```python
# Call the model on some test data to initialize it
model.predict(x_test[:1])  # Make a prediction on a single test image to initialize the model
```

```python

import matplotlib.pyplot as plt
# Get model predictions
predictions = model.predict(x_test)

# Display the first few test images, predicted labels, and true labels
for i in range(5):
    plt.imshow(x_test[i].reshape(28, 28), cmap='gray')
    plt.title(f"True Label: {np.argmax(y_test[i])}, Predicted: {np.argmax(predictions[i])}")
    plt.show()
```

```python
# Evaluate the model on the test data
test_loss, test_accuracy = model.evaluate(x_test, y_test, verbose=0)

# Print the accuracy
print(f"Test Accuracy: {test_accuracy * 100:.2f}%")
```
